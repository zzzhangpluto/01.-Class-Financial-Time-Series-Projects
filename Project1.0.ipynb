## Rationality inside Machine Learning methods

1)Measurement of an asset’s risk premium is fundamentally a problem of prediction—the risk premium is the conditional expectation of a future realized excess return. Machine learning, whose methods are largely specialized for prediction tasks, is thus ideally suited to the problem of risk premium measurement.  
2)Variables are too large for traditional prediction methods.  
3)Machine learning can handle the non-linear and high-dimension problem.

## Method

1) OLS as a benchmark  
2) Principal components regression (PCR) and partial least squares (PLS) to reduce dimension.
